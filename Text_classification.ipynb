```python
import pandas as pd

df = pd.read_csv("/content/ecommerceDataset.csv")
df.columns= ["catagory","discription"]
print(df.tail())
print(df.shape)
```

```python
df.catagory.value_counts()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.catagory.replace({"Clothing & Accessories":"Clothing"},inplace=True)
df.catagory.unique()

```

```python
df['catagory'] = "__label__" + df['catagory'].astype(str)
df.head()
```

```python
df['catagory_discription'] = df['catagory'] +" "+df['discription']
df.head()
```

```python
df['catagory_discription'][6]
```

```python
import re

def preproces(text):
  text = re.sub('[^\s\w\']', ' ',text)
  text = re.sub(' +',' ',text).lower()
  return text
```

```python
preproces(" VIKI's | Bookcase/Bookshelf   (3-shelf/shelve,  white) | ?/  ,  hi")
```

```python
df['catagory_discription'] = df['catagory_discription'].map(preproces)
df.head(4)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df,test_size=0.2)
test.shape , train.shape
```

```python
test.head(3)
```

```python

train.to_csv("ecommerceDataset.test",columns = ["catagory_discription"],index = False ,header = False)
test.to_csv("ecommerceDataset.train",columns = ["catagory_discription"],index = False ,header = False)
```

```python
import fasttext

model= fasttext.train_supervised(input = 'ecommerceDataset.train')
model.test("ecommerceDataset.test")
```
